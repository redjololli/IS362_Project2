```python
# Renaming the columns
crime.rename(columns = {'Table 4': 'State', 'Unnamed: 1': 'City', 'Unnamed: 2': 'Year', 'Unnamed: 3': 'Population',
                          'Unnamed: 4': 'Violent Crime', 'Unnamed: 5': 'Murder', 'Unnamed: 6': 'Rape',
                          'Unnamed: 7': 'Robbery', 'Unnamed: 8': 'Aggravated Assault', 'Unnamed: 9': 'Property Crime',
                          'Unnamed: 10': 'Burglary', 'Unnamed: 11': 'Larceny Theft', 'Unnamed: 12': 'Motor Vehicle Theft',
                          'Unnamed: 13': 'Arson'}, inplace=True)
```

```python
# Adding underscore in between spaces from column names
crime.rename(columns={'Violent Crime': 'Violent_Crime', 'Aggravated Assault': 'Aggravated_Assault',
                        'Property Crime': 'Property_Crime', 'Larceny Theft': 'Larceny_Theft',
                        'Motor Vehicle Theft': 'Motor_Vehicle_Theft'}, inplace=True)
```

```python
crime.columns
```

```python
# returns basic statistical details like percentile, mean, std etc. of a data frame or a series of numeric values. 
crime.describe()
```

```python
# Change NaNs to 0

crime.fillna(0, inplace = True)

crime.head(12)
```

```python
# Replacing all commans of the table
crime['Population'] = crime['Population'].str.replace(',', '')
crime['Violent_Crime'] = crime.Violent_Crime.str.replace(',', '')
crime['Murder'] = crime.Murder.str.replace(',', '')
crime['Rape'] = crime.Rape.str.replace(',', '')
crime['Robbery'] = crime.Robbery.str.replace(',', '')
crime['Aggravated_Assault'] = crime.Aggravated_Assault.str.replace(',', '')
crime['Property_Crime'] = crime.Property_Crime.str.replace(',', '')
crime['Burglary'] = crime.Burglary.str.replace(',', '')
crime['Larceny_Theft'] = crime.Larceny_Theft.str.replace(',', '')
crime['Motor_Vehicle_Theft'] = crime.Motor_Vehicle_Theft.str.replace(',', '')
crime['Arson'] = crime.Arson.str.replace(',', '')
```

```python
crime['Year'] = crime.Year.astype(int)
```

```python
#Changig data type from object to int, except STATE and CITY.
crime.fillna(0, inplace = True)
crime[['Population']] = crime[['Population']].astype(int)
crime['Violent_Crime'] = crime.Violent_Crime.astype(int)
crime['Murder'] = crime.Murder.astype(int)
crime['Rape'] = crime.Rape.astype(int)
crime['Robbery'] = crime.Robbery.astype(int)
crime['Aggravated_Assault'] = crime.Aggravated_Assault.astype(int)
crime['Property_Crime'] = crime.Property_Crime.astype(int)
crime['Burglary'] = crime.Burglary.astype(int)
crime['Larceny_Theft'] = crime.Larceny_Theft.astype(int)
crime['Motor_Vehicle_Theft'] = crime.Motor_Vehicle_Theft.astype(int)
crime['Arson'] = crime.Arson.astype(int)
```
